```python
from pymongo import MongoClient
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import pprint
```

```python
client = MongoClient()
capstone_db=client['capstone']
rating = capstone_db['femi']
```

```python
url = 'https://www.rottentomatoes.com/m/solis'
```

```python
r = requests.get(url)
r.status_code
```

```python

pprint.pprint(r.text)
```

```python
whole_page = capstone_db['whole_page']
```

```python
whole_page.insert_one({'html': r.content, 'url':url, 'time_scraped':time.time()})
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```
